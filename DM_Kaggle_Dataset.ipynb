#####  *heart.csv 데이터 속성 설명*
   * age        = 나이
   
   * sex        = 성별 (0<-여자, 1<-남자)
   
   * cp         = 흉통 유형 (1= 전형적인 협십증, 2= 비정형 협십증, 3= 비형심 통증 4= 무증상)
   
   * trestbps   = 안정혈압(병원 입원시  )
   
   * chol       = 혈청콜레스테롤
   
   * fbs        = 공복 혈당  (120mg 넘으면 true = 1 , 0 = false)
   
   * restecg    = 심전도 결과( 0= 정상 1= ST-T파 이상) 
   
   * thalach    = 최대 심박수 
   * exang      = 운동 유발 혐심증(1= true , 0 = false)
   * oldpeak    = ST depression(=상동맥이 허혈성 즉 혈관이 잘 흐르고 있지는 않다)
                  운동과 휴식에 관련 된 ST depression 
              
   * slope      =  st segment는 그래프에서 s-wave에서 t-wave까지의 구간을 의미
                   운동시 최대  slope of the peak 기울기
                   (1=상향기울기, 2=편평함 , 3 = 하양기울기)
            
   * ca         =  혈관 투시에 의한 채색된 주요 혈관수 (0~3으로 표현)  
   * thal       =  3=정상 , 6 =고정된 결함  7 = 가역 결함
   * target     = 0과 1로 표현

#### 40대의 심장병 추출
   * --기준--
   1. 40 ~ 49 사이의 남녀 
   2. cp(흉통)이 값이1(전형적인  협십증) 인경우
   3. restecg(운동 유발 협심증)의 값이  1(ST-T파 이상)인경우
   4. thal값이 6(고정된 결합)인 경우
   5. slop값이 1(상향 기울기),3(하양 기울이)인 경우 ( st segment구간에 상승 하강 있는 경우 )
   6. 1 ~ 5이 모두 포함된 사람 추출
   

```python
 heart.loc[((heart['age']>=40) & (heart['age']<=49)) & (heart['cp']==1) & (heart['restecg']==1) & (heart['thal']==1)
            & (heart['slope']==1) | (heart['slope']==3 ),: ]
```

#### (10) 40미만 연령에 대한 개수, 평균 확인
 * loc를 활용하여 age값이 40미만의 행만을 추출.
 * 모든 사람은 나이를 갖고 있으니 age행 개수로 40대 미만 사람명 수 추출.
 * mean()를 통해서 40대 미만 평균을 추출.

```python
h1 = heart.loc[heart['age']<40,:]

print("40대 미만 사람 수 = ",h1['age'].count())

print("40대 미만 평균    = ",h1.mean())
```

#### (10) 40-55 연령에 대한 개수, 평균 확인 
 * loc를 활용하여 age값이 40 ~ 55의 행만을 추출.
 * 모든 사람은 나이를 갖고 있으니 age행 개수로 40~ 55사이의 사람명 수 추출.
 * mean()를 통해서 40~ 50 사이의 평균을 추출.

```python
h2 = heart.loc[(heart['age']>=40) & (heart['age']<=55) ,:]

print("40~ 55 사이의 사람 수 = ",h2['age'].count())
print("40~ 55 사이의 평균    =  ",h2.mean())
```

#### (10) 50초과 연령에 대한 개수, 평균 확인

 * 모든 사람은 나이를 갖고 있으니 age행 개수로 50대 초과 사람명 수 추출.
 * mean()를 통해서 50대 초과 평균을 추출.

```python
h3 = heart.loc[(heart['age']>50) ,:]

print("05 사이의 사람 수 = ",h3['age'].count())
print("40~ 55 사이의 평균    =  ",h3.mean())
```
