### 3)데이터 추출

#### (3)폭 속성, 길이 속성만 추출

```python

sw  =  { "Sepal width": pd.Series( irisdata['Sepal width'] ) ,
         'Petal width': pd.Series( irisdata['Petal width'] )
          }

swdf = pd.DataFrame(sw)




sl   =  {"Sepal length": pd.Series( irisdata['Sepal length'] ),
        "Petal length": pd.Series( irisdata['Petal length'] )
        }
sldf = pd.DataFrame(sl)


display(swdf)
display(sldf)

```

### 4)데이터 추가

#### (1)꽃받침 길이( Sepal length)와 꽃받침 폭( Sepal width)의 비율을 새로운 속성으로 추가

```python
irisdata['Sepal Ratio'] = irisdata['Sepal length'] / irisdata['Sepal width']
irisdata
```

#### (2)꽃잎 길이(Petal length) 와 꽃잎 폭(Petal width)의 비율을 새로운 속성으로 추가

```python
irisdata['Petal Ratio'] = irisdata['Petal length'] / irisdata['Petal width']
irisdata
```
